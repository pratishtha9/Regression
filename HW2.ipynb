sm.add_constant(X) is used to add the constant values in the table.

R-squared value gives the fraction of variation in the response that is predicted by the predictors. The high value of R-squared which is almost as close to the ideal R squared value (1) indicates that the data points are very close to the regression line and there is very less error(RSS) in prediction. 

The value of F statistic shows that there is some relationship between the predictors and the response.

t value is the test statistic values used to test the Null Hypothesis. 

t=(coef-0)/std err with n-2 degrees of freedom, where n is the sample size.

Confidence Interval is the 95% probablity of the unknown paramater(or the coefficients we are trying to estimate) being in this range:-βˆ1 ± 2 ·SE(βˆ1). In this case the confidence interval is from [0.025 to 0.975].

We reject the null hypothesis when the value of p is very small, i.e. lesser than 0.05 because we usually take 5% level of significance. We can reject the null hypothesis for all the predictors because for all of them the p value is 0. This means that all the predictors have some relationship with the response variable.
